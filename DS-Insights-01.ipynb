# Análise de dados - Lojas Contoso


<h1 style="color: red">Coleta de dados</h1>

### 1.1 - Importando bibliotecas

```python
import pandas as pd
import matplotlib as plt
```

### 1.2 - Importando a base de dados

```python
# Importanto (lembrar do encoding caso precise)
vendas_df = pd.read_csv("Dataset - Contoso/Contoso - Vendas - 2017.csv", sep=";")
produtos_df = pd.read_csv("Dataset - Contoso/Contoso - Cadastro Produtos.csv", sep=";")
lojas_df = pd.read_csv("Dataset - Contoso/Contoso - Lojas.csv", sep=";")
clientes_df = pd.read_csv("Dataset - Contoso/Contoso - Clientes.csv", sep=";")

# Exibindo dataframes
display(vendas_df)
display(produtos_df)
display(lojas_df)
display(clientes_df)

```

<h1 style="color: red">Limpeza de dados</h1>

### Filtrando colunas que interessam
Vamos pegar das tabelas principais somente as colunas que nos interessam.<br>
A limpeza de dados pode ser feita "pegando somente os dados que interessam" ou removendo os dados que não interessam (método drop).

```python
# Selecionando as colunas
clientes_df = clientes_df[['ID Cliente', "E-mail"]]
produtos_df = produtos_df[['ID Produto', "Nome do Produto"]]
lojas_df = lojas_df[['ID Loja', "Nome da Loja"]]
# Mesclando as 3 tabelas em uma só
# Lembrando que só mescla se tiverem colunas com mesmo nome em ambas
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')
display(vendas_df.head(3))
#vendas_df.to_csv("df_filtrada.csv", sep=";")
```

### Procurando por valores nulos

```python
vendas_df.isnull().sum()
# Note que nenhuma coluna possui valores nulos
```

### Verificando se os tipos de dados estão corretos
*Se valores numéricos são numéricos, string são objects e  etc
caso contrario, nessa etapa devemos converte-los.

```python
vendas_df.info()
```

### Convertendo dados
A coluna de data está formatada como object, vamos converter para data (to_datetime())<br>
Vamos também separar a data em colunas de mes e ano.

```python
# Pandas ja fornece uma ferramenta de conversão, temos que informar a "formatação atual" da data
vendas_df['Data da Venda'] = pd.to_datetime(vendas_df['Data da Venda'], format='%d/%m/%Y')
# Vamos separar a data em colunas de mes e ano  ( se coluna nao existe pandas cria)
vendas_df['Ano da Venda'] = vendas_df['Data da Venda'].dt.year
vendas_df['Mes da Venda'] = vendas_df['Data da Venda'].dt.month
```

<h1 style="color: red">Exploração dos dados</h1>

### Qual o cliente comprou mais vezes?

```python
# value_counts() conta quants vezes um item aparece numa coluna
top_clientes = vendas_df["E-mail"].value_counts()
# plota, tamanho e passo do eixo Y
top_clientes[:3].plot(figsize=(10, 4), yticks=range(67, 80, 1))
```

### Qual a loja que mais vendeu?

```python
# groupbytransforma nome da loja em índice e soma os numéricos sum()
top_vendas = vendas_df.groupby("Nome da Loja").sum()
# Filtra somente a coluna quantidade vendida
top_vendas = top_vendas[["Quantidade Vendida"]]
# Coloca a coluna em ordem decrescente
top_vendas = top_vendas.sort_values("Quantidade Vendida", ascending=False)
display(top_vendas)
top_vendas[:10].plot(figsize=(10,5), kind="bar")

# melhor desempenho max() idxmax()
melhor_venda = top_vendas["Quantidade Vendida"].max()
melhor_loja = top_vendas["Quantidade Vendida"].idxmax()

# pior desempenho min() idxmin()
pior_venda = top_vendas["Quantidade Vendida"].min()
pior_loja = top_vendas["Quantidade Vendida"].idxmin()

print(f"Melhor desempenho: {melhor_loja} vendeu: {melhor_venda:,.2f}")
print(f"Pior desempenho: {pior_loja} vendeu: {pior_venda:,.2f}")

print(f'Quantidade de lojas: {vendas_df["Nome da Loja"].nunique()}')
```

### Qual o volume de vendas e a porcentagem de devoluçao de produtos?

```python
vol_vendas = vendas_df["Quantidade Vendida"].sum()
vol_devol = vendas_df["Quantidade Devolvida"].sum()

print(f"Volume de vendas : {vol_vendas:,.2f}")
print(f"Volume devoluções: {vol_devol:,.2f}")
print(f"Porcentagem de devoluções: {vol_devol/vol_vendas:.2%}")
```

### Qual a porcentagem de devolução da loja que mais vende?

```python
# Filtrando a melhor loja pelo nome da loja = Loja Contoso Catalog
dev_melhor_loja = vendas_df[vendas_df["Nome da Loja"] == "Loja Contoso Catalog "]
qnt_vend = dev_melhor_loja["Quantidade Vendida"].sum()
qnt_dev = dev_melhor_loja["Quantidade Devolvida"].sum()

# Filtrando a pior loja pelo nome da loja = Loja Contoso Venezia 
dev_pior_loja = vendas_df[vendas_df["Nome da Loja"] == "Loja Contoso Venezia "]
qnt_vendp = dev_pior_loja["Quantidade Vendida"].sum()
qnt_devp = dev_pior_loja["Quantidade Devolvida"].sum()

print(f"A taxa de devolução da melhor loja ({melhor_loja}) é: {qnt_dev/qnt_vend:.2%}")
print(f"A taxa de devolução da pior loja ({pior_loja}) é: {qnt_devp/qnt_vendp:.2%}")
```

### Quais os melhores e piores meses de vendas?

```python
vendas_data = vendas_df[["Nome da Loja", "Mes da Venda", "Quantidade Vendida"]]
vendas_data = vendas_data.groupby('Mes da Venda').sum()
display(vendas_data.sort_values("Quantidade Vendida", ascending=False))
vendas_data.plot(figsize=(10,5), kind="bar")
vendas_data.sort_values("Quantidade Vendida", ascending=False).plot(figsize=(10,5), kind="bar")
```

### Qual produto mais vendido e menos vendido?

```python
vendas_prod = vendas_df[["Nome do Produto", "Quantidade Vendida"]]
vendas_prod = vendas_prod.groupby('Nome do Produto').sum()
display(vendas_prod.sort_values("Quantidade Vendida", ascending=False))

vl_mais_vendido = vendas_prod['Quantidade Vendida'].max()
prod_mais_vendido = vendas_prod['Quantidade Vendida'].idxmax()

vl_menos_vendido = vendas_prod['Quantidade Vendida'].min()
prod_menos_vendido = vendas_prod['Quantidade Vendida'].idxmin()

print(f"{prod_mais_vendido} vendeu {vl_mais_vendido:,.2f} unidades")
print(f"{prod_menos_vendido} vendeu {vl_menos_vendido:,.2f} unidades")
```
